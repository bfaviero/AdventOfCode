```python
# Part 1
# RPS -> 012
R, P, S = 0, 1, 2
them_map = "ABC"
my_map = "XYZ"
move_points_map = [1, 2, 3]
win_points_map = [3, 6, 0]
winmap = [
    [0, -1, 1],
    [1, 0, -1],
    [-1, 1, 0]
]

# X = lose, Y = draw, Z = win
outcome_map = [-1, 0, 1]

total_points = 0
for line in lines:
    (move1, outcome_symbol) = line.split(" ")
    theirs = them_map.find(move1)
    outcome = outcome_map[my_map.find(outcome_symbol)]
    for mine in range(len(winmap)):
        if winmap[mine][theirs] == outcome:
            break
    move_points = move_points_map[mine]
    win_points = win_points_map[winmap[mine][theirs]]
    total_points += move_points + win_points
total_points
```
