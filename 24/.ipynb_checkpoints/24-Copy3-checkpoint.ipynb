```python
height = len(mapp)
width = len(mapp[0])

starting_point = (0, mapp[0].find("."))
start_row = 0
start_col = mapp[0].find(".")
end_row = height - 1
end_col = mapp[-1].find(".")


transforms = {
    None: lambda row, col: (row, col),
    "^": lambda row, col: (row - 1, col),
    "v": lambda row, col: (row + 1, col),
    "<": lambda row, col: (row, col - 1),
    ">": lambda row, col: (row, col + 1),
}

wrapmap = {
    "^": lambda row, col: (height - 2, col),
    "v": lambda row, col: (1, col),
    "<": lambda row, col: (row, width - 2),
    ">": lambda row, col: (row, 1),
}


def transform(row, col, direction):
    return transforms[direction](row, col)


def off_limits(row, column):
    if (row, column) == (start_row, start_col):
        return False
    if (row, column) == (end_row, end_col):
        return False
    return row <= 0 or row >= height - 1 or column <= 0 or column >= width - 1
        

def get_next_position(mover, wrap=True):
    (row, col, direction) = mover
    row, col = transform(row, col, direction or direction)
    if off_limits(row, col) and wrap:
        row, col = wrapmap[direction](row, col)
    return row, col

def get_next_position_with_direction(mover, wrap=True):
    row, col = get_next_position(mover, wrap)
    return (row, col, mover[2])

        
    
def move_to_next_position(mover, wrap=True):
    (row, col, direction) = mover
    row, col = get_next_position((row, col, direction, wrap))
    return (row, col, direction)




elements = []
for row in range(height):
    for col in range(width):
        val = mapp[row][col]
        if val not in {".", "#"}:
            elements.append((row, col, val))
        


def print_points(elements, elf_row, elf_col):
    points = {(b[0], b[1]): b for b in elements}
    for row in range(height):
        for col in range(width):
            if (row, col) in points:
                (_, _, direction) = points[(row, col)]
                print(direction, end="")
            elif (row, col) == (elf_row, elf_col):
                print("%", end="")
            elif row == 0 or row == height - 1 or col == 0 or col == width - 1:
                print("#", end="")
            else:
                print(".", end="")
        print("")

def get_valid_moves(elements, new_blizzard_points, elf_row, elf_col):
    moves = [transform(elf_row, elf_col, d) for d in "^v<>"] + [(elf_row, elf_col)]
    valid_moves = [move for move in moves if move not in new_blizzard_points and not off_limits(move[0], move[1])]
    return valid_moves


state = {}
positions = elements
area = (width - 2) * (height - 2)
for rd in range(area):
    state[rd] = positions
    positions = set(get_next_position_with_direction(elt) for elt in positions)
    
    

things_to_try = [(start_row, start_col, 0)]
def run(row, col, end):
    hashmap = set()
    while things_to_try:  
        (row, col, rd) = things_to_try.pop(0)
        key = "{},{},{}".format(row, col, rd)
        if key in hashmap:
            continue
        else:
            hashmap.add(key)

        positions = state[(rd) % area]
        new_positions = state[(rd+1) % area]
        valid_moves = get_valid_moves(positions, new_positions, row, col)

        if not valid_moves:
            continue


        for (row, col) in valid_moves:
            things_to_try.append((row, col, (rd + 1) % area))
            
        if end in valid_moves:
            print("Found: {} ({} moves)".format(end, rd))
            return (rd, new_positions)
            break


new_positions = run(start_row, start_col, (end_row, end_col))
rd2, new_positions = run(end_row, end_col, (start_row, start_col))
rd3, new_positions = run(start_row, start_col, (end_row, end_col))


```
