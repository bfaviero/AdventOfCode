```python
# Part 1
from collections import deque
lines = crate_data.split("\n")
num_line = lines[-1]
crates = [[]]
for column, num_string in enumerate(num_line):
    if num_string.strip():
        crates.append(deque([line[column] for line in lines[:-1] if line[column].strip()]))
        
        
import re
moves = move_data.split("\n")
for move in moves:
    num, from_idx, to_idx = list(map(int, re.findall("(\d+)", move)))
    for i in range(num):
        crates[to_idx].appendleft(crates[from_idx].popleft())
# print_crates()
print(''.join([crate[0] for crate in crates[1:] if crate]))
```

```python
def popleft(q, n):
    popped = []
    for i in range(n):
        popped.append(q.popleft())
    return deque(popped)
```

```python
# Part 2
from collections import deque
lines = crate_data.split("\n")
num_line = lines[-1]
crates = [[]]
for column, num_string in enumerate(num_line):
    if num_string.strip():
        crates.append(deque([line[column] for line in lines[:-1] if line[column].strip()]))
        
        
import re
moves = move_data.split("\n")
for move in moves:
    num, from_idx, to_idx = list(map(int, re.findall("(\d+)", move)))
    popped = popleft(crates[from_idx], num)
    crates[to_idx] = popped + crates[to_idx]
# print_crates()
print(''.join([crate[0] for crate in crates[1:] if crate]))
```
